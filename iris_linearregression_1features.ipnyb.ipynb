```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Loading the Iris dataset
data = load_iris()
print(data)
X = data.data
print(X)

y = data.target
print(y)

# Using only the first feature for a simple linear regression
X_simple = X[:, 0].reshape(-1, 1)

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_simple, y, test_size=0.3, random_state=42)

# Creating and training the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predicting on the test set
y_pred = model.predict(X_test)

# Plotting the scatter plot and the regression line
plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(X_test, y_test, color='blue', label='Actual Data')
plt.show()
plt.scatter(X_test, y_pred, color='blue', label='Actual Data')
plt.show()
# Regression line
plt.plot(X_test, y_test, color='red', linewidth=2, label='Regression Line')
plt.show()
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Regression Line')
plt.show()
plt.title('Linear Regression: Actual vs Predicted')
plt.xlabel('Feature Value')
plt.ylabel('Target Value')
plt.legend()
plt.show()
```
