```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv(r'D:\Drive D\مجتمع فنی تهران\machine learning\ترم خودم\6\clusttering.exercise\2spiral.csv')
df
```

```python
k=3
Xx=df[["a0", "a1"]]
model=KMeans(n_clusters=k, random_state= 42)
y_pred=model.fit_predict(Xx)
```

```python
Xx
```

```python
y_pred
```

```python
plt.scatter (df['a0'],df['a1'], c= y_pred)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['a0'], df['a1'],  c=y_pred, cmap='viridis', marker='o')

ax.set_xlabel('a0 Axis')
ax.set_ylabel('a1 Axis')

ax.set_title('3D KMeans Clustering')

plt.show()
```

```python
from sklearn.cluster import DBSCAN
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv(r'D:\Drive D\مجتمع فنی تهران\machine learning\ترم خودم\6\clusttering.exercise\2spiral.csv')
df
```
