```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# تنظیم استایل برای زیبایی نمودار
sns.set_style("whitegrid")

# بارگذاری داده‌ها
penguins = pd.read_csv("C:/Users/A.S.T.co/Downloads/penguins.csv")
print(penguins)

# انتخاب ستون‌های مورد نظر و حذف مقادیر گمشده
columns = ["Flipper Length (mm)", "Body Mass (g)", "Species", "Culmen Length (mm)", "Culmen Depth (mm)"]
penguins_non_missing = penguins[columns].dropna()
print(columns)
print(penguins_non_missing)
# تعریف X و y برای مدل رگرسیون
X = penguins_non_missing[["Flipper Length (mm)", "Culmen Length (mm)", "Culmen Depth (mm)"]]  # متغیرهای مستقل
y = penguins_non_missing["Body Mass (g)"]  # متغیر وابسته
print(X)
print(y)
# ایجاد و آموزش مدل رگرسیون خطی
model = LinearRegression()
model.fit(X, y)

# پیش‌بینی مقادیر
y_pred = model.predict(X)
print(y_pred)
# محاسبه دقت مدل (R²)
r2 = r2_score(y, y_pred)
print(f"R² Score: {r2:.4f}")

# نمایش ضرایب مدل
print("ضرایب مدل:", model.coef_)
print("عرض از مبدا:", model.intercept_)

# رسم نمودار پراکندگی و خط رگرسیون برای یکی از متغیرهای مستقل
plt.figure(figsize=(8, 5))
sns.scatterplot(x=penguins_non_missing["Flipper Length (mm)"], y=y, label="real")
sns.lineplot(x=penguins_non_missing["Flipper Length (mm)"], y=y_pred, color='red', label="predicted")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.title("linear regression of Flipper Length (mm) and Body Mass (g)")
plt.legend()
plt.show()

# رسم نمودار پراکندگی و خط رگرسیون برای یکی از متغیرهای مستقل
plt.figure(figsize=(8, 5))
sns.scatterplot(x=penguins_non_missing["Culmen Length (mm)"], y=y, label="real")
sns.lineplot(x=penguins_non_missing["Culmen Length (mm)"], y=y_pred, color='red', label="predicted")
plt.xlabel("Culmen Length (mm)")
plt.ylabel("Body Mass (g)")
plt.title("linear regression of Culmen Length (mm) and Body Mass (g)")
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(x=penguins_non_missing["Culmen Depth (mm)"], y=y, label="real")
sns.lineplot(x=penguins_non_missing["Culmen Depth (mm)"], y=y_pred, color='red', label="predicted")
plt.xlabel("Culmen Depth (mm)")
plt.ylabel("Body Mass (g)")
plt.title("linear regression of Culmen Depth (mm) and Body Mass (g)")
plt.legend()
plt.show()


# رسم Pairplot از ویژگی‌ها
sns.pairplot(penguins_non_missing[columns])
plt.suptitle("Pairplot of Features", y=1.02)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# تنظیم استایل نمودار
sns.set_style("whitegrid")

# بارگذاری داده‌ها
file_path = "C:/Users/A.S.T.co/Downloads/penguins.csv"
penguins = pd.read_csv(file_path)

# بررسی نام ستون‌ها
print("ستون‌های دیتاست:", penguins.columns)

# انتخاب ستون‌های مورد نیاز و حذف مقادیر گمشده
columns = ["Flipper Length (mm)", "Body Mass (g)", "Culmen Length (mm)", "Culmen Depth (mm)"]
penguins_non_missing = penguins[columns].dropna()

# نمایش اطلاعات اولیه پس از حذف مقادیر گمشده
print("اطلاعات پس از حذف مقادیر گمشده:\n", penguins_non_missing.head())

# تعریف X و y برای مدل
X = penguins_non_missing[["Flipper Length (mm)", "Culmen Length (mm)", "Culmen Depth (mm)"]]
y = penguins_non_missing["Body Mass (g)"]

# ایجاد و آموزش مدل رگرسیون خطی
model = LinearRegression()
model.fit(X, y)

# پیش‌بینی مقدار y
y_pred = model.predict(X)

# محاسبه R²
r2 = r2_score(y, y_pred)
print(f"R² Score: {r2:.4f}")

# نمایش ضرایب مدل
print("ضرایب مدل:", model.coef_)
print("عرض از مبدا:", model.intercept_)

# رسم نمودار پراکندگی و خط رگرسیون برای هر متغیر مستقل
features = ["Flipper Length (mm)", "Culmen Length (mm)", "Culmen Depth (mm)"]
titles = [
    "Linear Regression: Flipper Length vs. Body Mass",
    "Linear Regression: Culmen Length vs. Body Mass",
    "Linear Regression: Culmen Depth vs. Body Mass"
]

for i, feature in enumerate(features):
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=X[feature], y=y, label="real")
    sns.lineplot(x=X[feature], y=y_pred, color='red', label="predicted")
    plt.xlabel(feature)
    plt.ylabel("Body Mass (g)")
    plt.title(titles[i])
    plt.legend()
    plt.show()


# رسم Pairplot از ویژگی‌ها
sns.pairplot(penguins_non_missing[columns])
plt.suptitle("Pairplot of Features", y=1.02)
plt.show()
```
